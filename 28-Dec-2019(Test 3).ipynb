## 2)Combine both the test and train datasets.

```python
merge_data = pd.merge(train_data,test_data,on='Self_Employed')
merge_data
```

## 3.Convert Gender,Married,Dependents,Education,Self_Employed,Property_Area into numeric form depending on column?

```python
m3 = merge_data
m3
m3['Gender_x'].replace('Male',0,inplace=True)
m3['Gender_x'].replace('FeMale',1,inplace=True)
m3['Married_x'].replace('No',0,inplace=True)
m3['Married_x'].replace('Yes',1,inplace=True)
m3['Dependents_x'].replace('3+',3,inplace=True)
m3['Education_x'].replace('Graduate',1,inplace=True)
m3['Education_x'].replace('Not Graduate',0,inplace=True)
m3['Self_Employed'].replace('No',0,inplace=True)
m3['Self_Employed'].replace('Yes',1,inplace=True)
m3['Property_Area_y'].replace('Urban',1,inplace=True)
m3['Property_Area_y'].replace('Semi Urban',2,inplace=True)
m3['Property_Area_y'].replace('Rural',3,inplace=True)
```
